```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("image_open_close.jpg",0)
image=cv2.cvtColor(image,cv2.COLOR_GRAY2RGB)
```

```python
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
kernel=cv2.getStructuringElement(cv2.MORPH_CROSS,(11,11))
```

```python
image2=cv2.morphologyEx(image,cv2.MORPH_CLOSE,kernel)
image1=cv2.morphologyEx(image,cv2.MORPH_OPEN,kernel)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original")
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Opening")
plt.imshow(image1)
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original")
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Closing")
plt.imshow(image2)
plt.axis("off")
plt.show()
```
